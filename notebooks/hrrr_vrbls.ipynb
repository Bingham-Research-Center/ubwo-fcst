```python
def setup_herbie(inittime, fxx=0, product="nat", model="hrrr"):
    H = Herbie(
    inittime,
    model=model,
    product=product,
    fxx=fxx,
    )
    return H
```

```python
inittime = "2023-11-29 18:00"
fxx = 6
H = setup_herbie(inittime, fxx=fxx)
```
